```python
# Test of different ways of plot external textbox.
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Patch

# Sample data
data = np.random.rand(10, 10)

# Create the image
fig, ax = plt.subplots()
image = ax.imshow(data, cmap='viridis')

# Add colorbar
cbar = plt.colorbar(image, ax=ax)
cbar.set_label('Intensity')


# Define custom legend handles
legend_handles = [Patch(facecolor='blue', edgecolor='r', label='Label 1'),
                  Patch(facecolor='green', edgecolor='r', label='Label 2')]

# Adjust the subplot parameters to give some space for the external legend
plt.subplots_adjust(right=0.8)

# Add the external legend
legend = ax.legend(handles=legend_handles, loc='upper left', bbox_to_anchor=(1, 1))

# Show the plot
plt.show()
```
